```python
def title(name):
    r_surname = name.split(',')[1]
    title = r_surname.split('.')[0]
    title = title.strip().lower()
    return title
```

```python
def title(name):
    title_group ={'mr':'mr',
                   'mrs':'mrs',
                   'miss':'miss',
                   'master':'master',
                   'don':'sir',
                   'rev':'sir',
                   'dr':'officer',
                   'mme':'mrs',
                   'ms':'mrs',
                   'major':'officer',
                   'lady':'lady',
                   'sir':'sir',
                   'mlle':'miss',
                   'col':'officer',
                   'capt':'officer',
                   'the countess':'lady',
                   'jonkheer':'sir',
                   'dona':'lady'}
    r_surname = name.split(',')[1]
    title = r_surname.split('.')[0]
    title = title.strip().lower()
    return title_group[title]
```

```python
df['Name'].map(lambda x : title(x)).unique()
```

```python
df['title'] = df['Name'].map(lambda x : title(x))
```

```python
df["IsMale"] = np.where((df.Sex == 'male'),1,0)
```

```python
df = pd.get_dummies(df,columns =['Pclass','title','Fare_category','Embarked','AgeState'])
```

```python
def get_cabin(cabin):
    return np.where(pd.notnull(cabin),str(cabin)[0].lower(),'z')
df['Deck'] = df['Cabin'].map(lambda x : get_cabin(x))
```
